```python
import pandas as pd
data = pd.read_csv('data/iris.csv')
data = data.dropna()
pd.concat([data[:10], data[60:70], data[110:120]])
```

# Sepal i Petal

```python
data.describe()
```

```python
data.groupby('class').describe()
```

```python
data[['sepal length', 'sepal width', 'petal length', 'petal width']].hist()
```

```python

parameters = data.columns.tolist()
parameters.remove('class')
pairs = [(x, y) for x in parameters for y in parameters if x != y]
for pair in pairs:
    ax = data[data['class'] == 'Iris-setosa'].plot(kind='scatter', x=pair[0], y=pair[1], label='setosa', color='DarkBlue');
    bx = data[data['class'] == 'Iris-versicolor'].plot(kind='scatter', x=pair[0], y=pair[1],label='versicolor', color='DarkGreen', ax= ax);
    cx = data[data['class'] == 'Iris-virginica'].plot(kind='scatter', x=pair[0], y=pair[1], label='virginica', color='Red', ax=bx);
```
